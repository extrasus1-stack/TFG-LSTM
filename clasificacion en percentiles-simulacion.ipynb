```python
import numpy as np

# ==============================================================================
# MÓDULO DE DECISIÓN CLÍNICA (SOPORTE A LA DECISIÓN - CDSS)
# ==============================================================================

# 1. SIMULACIÓN DE DATOS HISTÓRICOS (Tus 420 pacientes ya entrenados)
# Simulamos que la IA les dio un Score de Riesgo medio de 0.50 a la mayoría.
np.random.seed(42) # Para que siempre salga el mismo resultado
scores_historicos = np.random.normal(loc=0.50, scale=0.15, size=420)

# Calculamos los umbrales (Percentil 25 y 75)
umbral_bajo = np.percentile(scores_historicos, 25)
umbral_alto = np.percentile(scores_historicos, 75)

print("📊 RESUMEN DE TU HOSPITAL (420 Pacientes):")
print(f"-> Umbral de Bajo Riesgo (P25): por debajo de {umbral_bajo:.3f}")
print(f"-> Umbral de Alto Riesgo (P75): por encima de {umbral_alto:.3f}\n")

# 2. FUNCIÓN DE RECOMENDACIÓN CLÍNICA
def generar_informe(id_paciente, score_paciente):
    print("=" * 60)
    print(f"🏥 INFORME IA - PACIENTE: {id_paciente}")
    print(f"   Score de Riesgo Asignado: {score_paciente:.3f}")
    print("-" * 60)

    if score_paciente >= umbral_alto:
        print("⚠️ CLASIFICACIÓN: ALTO RIESGO de Recidiva / EMR")
        print("💡 RECOMENDACIÓN: ADELANTAR control analítico.")
        print("   Sugerencia: Monitorización estrecha (ej. cada 15 días) y ")
        print("   valoración de intervención temprana.")

    elif score_paciente <= umbral_bajo:
        print("✅ CLASIFICACIÓN: BAJO RIESGO")
        print("💡 RECOMENDACIÓN: Posibilidad de RETRASAR/ESPACIAR control.")
        print("   Sugerencia: Evolución favorable. Seguro ampliar intervalos.")

    else:
        print("⚖️ CLASIFICACIÓN: RIESGO INTERMEDIO")
        print("💡 RECOMENDACIÓN: MANTENER pauta fija.")
        print("   Sugerencia: Seguir protocolo estándar (Día 30, 60, 90...).")
    print("=" * 60 + "\n")

# 3. SIMULACIÓN EN CONSULTA (Llegan 3 pacientes nuevos)
pacientes_nuevos = {
    "Carlos (Caso 1)": 0.85,  # Le simulamos un score muy alto
    "Ana (Caso 2)": 0.48,     # Le simulamos un score medio
    "Luis (Caso 3)": 0.20     # Le simulamos un score muy bajo
}

for nombre, score in pacientes_nuevos.items():
    generar_informe(nombre, score)
```
