```python
import numpy as np

# ==============================================================================
# MÓDULO DE DECISIÓN CLÍNICA (SOPORTE A LA DECISIÓN - CDSS)
# ==============================================================================

# 1. SIMULACIÓN DE DATOS HISTÓRICOS (Tus 420 pacientes ya entrenados)
# Simulamos que la IA les dio un Score de Riesgo medio de 0.50 a la mayoría.
np.random.seed(42) # Para que siempre salga el mismo resultado
scores_historicos = np.random.normal(loc=0.50, scale=0.15, size=420)

# Calculamos los umbrales (Percentil 25 y 75)
umbral_bajo = np.percentile(scores_historicos, 25)
umbral_alto = np.percentile(scores_historicos, 75)

print("📊 RESUMEN DE TU HOSPITAL (420 Pacientes):")
print(f"-> Umbral de Bajo Riesgo (P25): por debajo de {umbral_bajo:.3f}")
print(f"-> Umbral de Alto Riesgo (P75): por encima de {umbral_alto:.3f}\n")

# 2. FUNCIÓN DE RECOMENDACIÓN CLÍNICA
def generar_informe(id_paciente, score_paciente):
    print("=" * 60)
    print(f"🏥 INFORME IA - PACIENTE: {id_paciente}")
    print(f"   Score de Riesgo Asignado: {score_paciente:.3f}")
    print("-" * 60)

    if score_paciente >= umbral_alto:
        print("⚠️ CLASIFICACIÓN: ALTO RIESGO de Recidiva / EMR")
        print("💡 RECOMENDACIÓN: ADELANTAR control analítico.")
        print("   Sugerencia: Monitorización estrecha (ej. cada 15 días) y ")
        print("   valoración de intervención temprana.")

    elif score_paciente <= umbral_bajo:
        print("✅ CLASIFICACIÓN: BAJO RIESGO")
        print("💡 RECOMENDACIÓN: Posibilidad de RETRASAR/ESPACIAR control.")
        print("   Sugerencia: Evolución favorable. Seguro ampliar intervalos.")

    else:
        print("⚖️ CLASIFICACIÓN: RIESGO INTERMEDIO")
        print("💡 RECOMENDACIÓN: MANTENER pauta fija.")
        print("   Sugerencia: Seguir protocolo estándar (Día 30, 60, 90...).")
    print("=" * 60 + "\n")

# 3. SIMULACIÓN EN CONSULTA (Llegan 3 pacientes nuevos)
pacientes_nuevos = {
    "Carlos (Caso 1)": 0.85,  # Le simulamos un score muy alto
    "Ana (Caso 2)": 0.48,     # Le simulamos un score medio
    "Luis (Caso 3)": 0.20     # Le simulamos un score muy bajo
}

for nombre, score in pacientes_nuevos.items():
    generar_informe(nombre, score)
```

```python
import numpy as np
from datetime import datetime, timedelta

umbral_bajo = 0.35
umbral_alto = 0.65

def evaluar_analitica_y_fechas(id_paciente, score_ia, ultima_analitica, fecha_analitica_hoy):
    print("=" * 65)
    print(f"🏥 INFORME CDSS - PACIENTE: {id_paciente}")
    print(f"   Score IA Asignado: {score_ia:.3f}")
    print(f"   Fecha muestra actual: {fecha_analitica_hoy.strftime('%d/%m/%Y')}")
    print("-" * 65)

    # Extraemos los valores (AHORA TODOS LOS '1' SON FACTORES DE RIESGO)
    wt1 = ultima_analitica['WT1']
    quim_malo = ultima_analitica['Quimerismo_Menor_95'] # 1 = <95% (Malo)
    mutacion = ultima_analitica['Mutacion']             # 1 = Sí (Malo)
    hla_dr_ausente = ultima_analitica['HLA_DR']         # 1 = No presenta (Malo), 0 = Sí presenta (Bueno)

    # --- CASO A: ALTO RIESGO ---
    if score_ia >= umbral_alto:
        proxima_cita = fecha_analitica_hoy + timedelta(days=15)
        print("⚠️ CLASIFICACIÓN: ALTO RIESGO (Adelantar control)")
        print(f"📅 PRÓXIMA ANALÍTICA EXACTA: {proxima_cita.strftime('%d/%m/%Y')} (en 15 días)")
        print("\n🔍 MOTIVOS DE ALARMA EN LA ÚLTIMA MUESTRA:")

        motivos = 0
        if wt1 == 1:
            print("   -> 🚩 Alerta: Detección de copias WT1 (<50 copias detectadas).")
            motivos += 1
        if quim_malo == 1:
            print("   -> 🚩 Alerta: Pérdida de quimerismo completo (< 95%).")
            motivos += 1
        if mutacion == 1:
            print("   -> 🚩 Alerta: Detección de mutaciones leucémicas post-TMO.")
            motivos += 1
        if hla_dr_ausente == 1:
            print("   -> 🚩 Alerta: Ausencia de HLA-DR (Pérdida de diana para tratamiento/mal pronóstico).")
            motivos += 1

        if motivos == 0:
            print("   -> ⚠️ Los valores dinámicos recientes son normales, pero la IA detecta")
            print("      riesgo latente por la combinación de variables basales e históricas.")

    # --- CASO B: BAJO RIESGO ---
    elif score_ia <= umbral_bajo:
        proxima_cita = fecha_analitica_hoy + timedelta(days=60)
        print("✅ CLASIFICACIÓN: BAJO RIESGO (Espaciar control)")
        print(f"📅 PRÓXIMA ANALÍTICA EXACTA: {proxima_cita.strftime('%d/%m/%Y')} (en 60 días)")
        print("\n🔍 MOTIVOS DE EVOLUCIÓN ÓPTIMA:")
        print("   -> ✔️ Quimerismo completo (100-95%) mantenido.")
        print("   -> ✔️ Ausencia de mutaciones leucémicas.")
        print("   -> ✔️ Enfermedad Mínima Residual (WT1) indetectable.")
        if hla_dr_ausente == 0:
            print("   -> ✔️ Presencia de HLA-DR (Células tratables / factor protector activo).")

    # --- CASO C: RIESGO INTERMEDIO ---
    else:
        proxima_cita = fecha_analitica_hoy + timedelta(days=30)
        print("⚖️ CLASIFICACIÓN: RIESGO INTERMEDIO (Mantener pauta)")
        print(f"📅 PRÓXIMA ANALÍTICA EXACTA: {proxima_cita.strftime('%d/%m/%Y')} (en 30 días)")
        print("\n🔍 ESTADO: Evolución estable. No hay marcadores críticos activos,")
        print("   pero el riesgo basal sugiere mantener el protocolo estándar.")

    print("=" * 65 + "\n")

# --- SIMULACIÓN DE LA CONSULTA MÉDICA ---
fecha_hoy = datetime.now()

# Paciente 1: Va mal (Pérdida de HLA-DR y mutación)
analitica_paciente1 = {'WT1': 0, 'Quimerismo_Menor_95': 0, 'Mutacion': 1, 'HLA_DR': 1}
evaluar_analitica_y_fechas("PAC_301", score_ia=0.82, ultima_analitica=analitica_paciente1, fecha_analitica_hoy=fecha_hoy)

# Paciente 2: Va perfecto (Presenta HLA-DR y todo lo demás a 0)
analitica_paciente2 = {'WT1': 0, 'Quimerismo_Menor_95': 0, 'Mutacion': 0, 'HLA_DR': 0}
evaluar_analitica_y_fechas("PAC_105", score_ia=0.21, ultima_analitica=analitica_paciente2, fecha_analitica_hoy=fecha_hoy)
```
